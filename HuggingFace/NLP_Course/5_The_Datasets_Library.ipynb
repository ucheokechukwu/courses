````python
import time
import math
from pathlib import Path
import pandas as pd
from tqdm.notebook import tqdm
````

````python
def fetch_issues(
    owner="huggingface",
    repo="datasets",
    num_issues=10_000,
    rate_limit=5_000,
    issues_path=Path('/content/drive/MyDrive/MLOPs_Projects/HuggingFace'),
    ):
    if not issues_path.is_dir():
        issues_path.mkdir(exist_ok=True)

    batch = []
    all_issues = []
    per_page = 100 # number of issues to return to page
    num_pages = math.ceil(num_issues / per_page)
    base_url = "https://api.github.com/repos"



    for page in tqdm(range(num_pages)):
        # query with state=all to get both open and closed issues
        query = f"issues?page={page}&per_page={per_page}&state=all"
        issues = requests.get(f"{base_url}/{owner}/{repo}/{query}",
                              headers=headers)
        batch.extend(issues.json())

        if len(batch) > rate_limit and len(all_issues) < num_issues:
            all_issues.extend(batch)
            batch = [] # flush batch for next time period
            print(f"Reached Github rate limit. Sleeping for an hour ....")
            time.sleep(60 + 1)

    all_issues.extend(batch)
    df = pd.DataFrame.from_records(all_issues)
    df.to_json(f"{issues_path}/{repo}-issues.jsonl",
               orient="records",
               lines=True)
    print(
        f"Downloaded all the issues for {repo}! Dataset stored at {issues_path/repo}-issues.jsonl"
    )
````

## Creating text embeddings

* https://www.sbert.net/examples/applications/semantic-search/README.html#symmetric-vs-asymmetric-semantic-search

````python
try:
    from transformers import AutoTokenizer, AutoModel
except:
    %pip install -q transformers
    from transformers import AutoTokenizer, AutoModel


model_ckpt = 'sentence-transformers/multi-qa-mpnet-base-dot-v1'
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
model = AutoModel.from_pretrained(model_ckpt)
````

````python
import torch
device = torch.device('cuda')
model.to(device)
````

````python
# we need to represent the issues corpus as a single vector
# so we need to pool or average our token embeddings
# one method is to use CLS pooling

def cls_pooling(model_output):
    return model_output.last_hidden_state[:, 0]

# helper funciton to:
# 1. tokenize a list of documents
# 2. place the tensors on the GPU
# 3. feed them into the model
# 4. apply CLS pooling to the outputs

def get_embeddings(text_list):
    # tokenize the documents
    encoded_input = tokenizer(
        text_list,
        padding=True,
        truncation=True,
        return_tensors='pt'
    )
    # place the tensors on the GPU
    encoded_input = {k:v.to(device) for k, v in encoded_input.items()}
    # feed them into the model
    model_output = model(**encoded_input)
    # apply CLS pooling to the outputs
    return cls_pooling(model_output)
````

````python
# testing with the first entry
embedding = get_embeddings(comments_dataset["text"][0])
embedding.shape
````

````python
embeddings_dataset = comments_dataset.map(
    lambda x: {"embeddings": get_embeddings(x['text']).detach().cpu().numpy()[0]}
)
````

## Using FAISS for efficient similarity search
The basic idea behind FAISS is to create a special data structure calledan index that allows one to find which embeddings are similar to an input embedding.

FAISS is a library that provides efficient algorithms to quickly search and cluster embedding vectors.

````python
# %pip install faiss-cpu faiss-gpu -q
# %pip install --upgrade faiss-cpu faiss-gpu -q
````

````python
embeddings_dataset.add_faiss_index(column='embeddings')
````

````python
# testing out query
question = "how can i load a dataset offline?"
question_embedding = get_embeddings([question]).cpu().detach().numpy()
question_embedding.shape
````

````python
# now we can compare to find the most similar embeddings
scores, samples = embeddings_dataset.get_nearest_examples(
    "embeddings", question_embedding, k=5
)
````

````python
import pandas as pd
samples_df = pd.DataFrame.from_dict(samples)
samples_df["scores"] = scores
samples_df.sort_values('scores', ascending=False, inplace=True)
samples_df
````

````python
# iterating over the first few rows to see if query matched the comments
for _, row in samples_df.iterrows():
    print(f"COMMENT: {row.comments}")
    print(f"SCORE: {row.scores}")
    print(f"TITLE: {row.title}")
    print(f"URL: {row.html_url}")
    print("=" * 50)
    print()
````
